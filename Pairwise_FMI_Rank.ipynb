This notebook consider the additional information of pairwise variables.  
Formula:
$$
Gain = F((X_1, X_2);Y) - F(X_1;Y) - F(X_2;Y)
$$
, where $F$ is reliable fraction of MI

As there are $1485$ pairwise variables, we choose equal frequency and set bin size as $\log_2(n)$, where $n$ is the sample sizes.

```python
import data1 as data
from common import *
import estimators
import pandas as pd
import itertools
from binning import *
from permutation import *
import math
```

```python
all_interactions = list(itertools.combinations([_ for _ in range(len(data.x.values[0]))], 2))
```

```python
k = int(math.log(data.x.shape[0]))
k
```

```python
result = []
for cell in all_interactions:
    X,X1,X2 = data.x.values[:, cell], data.x.values[:, cell[0]], data.x.values[:, cell[1]]
    col_name = np.array(data.x.columns)[list(cell)]
    res = gain(equal_freq(X, k), data.y.values) - Permutation(equal_freq(X, k), data.y.values).summary()/entropy(data.y.values)
    res1 = gain(equal_freq(X1, k), data.y.values) - Permutation(equal_freq(X1, k), data.y.values).summary()/entropy(data.y.values)
    res2 = gain(equal_freq(X2, k), data.y.values) - Permutation(equal_freq(X2, k), data.y.values).summary()/entropy(data.y.values)
    res = res-res1-res2
    result.append([res, " / ".join(col_name.tolist())])
```

```python
result.sort(reverse=True)
```

```python
df1 = pd.DataFrame({'Variable_Name': [each[1] for each in result], 'MI_scores':[each[0] for each in result]})
df1
```
